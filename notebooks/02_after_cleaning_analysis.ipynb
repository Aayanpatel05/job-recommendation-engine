```python
%pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
jobs = pd.read_csv('../tech_jobs_clean.csv')
```

```python
print(len(jobs))
print(list(jobs.columns))
```

```python
work_dist=jobs['formatted_work_type'].value_counts()
plt.figure(figsize=(10, 6))
work_dist.plot(kind='bar', color='lightcoral')
plt.title('Work Type Distribution')
plt.xlabel('Work Type')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
jobs['desc_length'] = jobs['description'].str.len()
print("Mean:", jobs['desc_length'].mean())
print("Median:", jobs['desc_length'].median())
print("Min:", jobs['desc_length'].min())
print("Max:", jobs['desc_length'].max())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(jobs['desc_length'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Job Description Lengths')
plt.xlabel('Description Length (characters)')
plt.ylabel('Frequency')
plt.axvline(jobs['desc_length'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.tight_layout()
plt.show()
```

```python
short_desc = jobs[jobs['desc_length'] < 400]
len(short_desc)
jobs = jobs[jobs["desc_length"] > 400]
```

```python
top_titles = jobs['title'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_titles.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.show()
```

```python
top_locations = jobs['location'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_locations.plot(kind='bar', color='lightblue')
plt.title('Top 10 Job Locations')
plt.xlabel('Location')
plt.ylabel('Number of Jobs')
plt.show()
```

```python
if 'remote_allowed' in jobs.columns:
    remote_count = jobs['remote_allowed'].sum()
    total = len(jobs)
    print(f"\nRemote-allowed jobs: {remote_count:,} ({remote_count/total*100:.1f}%)")
```

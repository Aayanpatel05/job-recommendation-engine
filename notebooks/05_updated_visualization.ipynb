```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
jobs = pd.read_csv('../data/processed/jobs_updated.csv')
pd.set_option('display.max_colwidth', None)
```

Explore Updated Jobs Skills Description

```python
has_skills = jobs['skills_desc'].notna().sum()
print(f"Percent of job descriptions with skills description: {has_skills / len(jobs) * 100:.2f}%")
```

```python
print("\nSample extracted skills (first 10 with skills):")
sample = jobs[jobs['skills_desc'].notna()].head(10)
for idx, row in sample.iterrows():
    print(f"\nTitle: {row['title']}")
    print(f"Skills: {row['skills_desc'][:100]}...")
```

```python
print("Average length of skills description:", jobs['skills_desc'].dropna().apply(len).mean())
print("Max length of skills description:", jobs['skills_desc'].dropna().apply(len).max())
print("Min length of skills description:", jobs['skills_desc'].dropna().apply(len).min())
print("Median length of skills description:", jobs['skills_desc'].dropna().apply(len).median())
```

```python
plt.figure(figsize=(10, 6))
jobs['skills_desc'].dropna().apply(len).hist(bins=30)
plt.xlabel('Length of Skills Description')
plt.ylabel('Frequency')
plt.title('Distribution of Skills Description Lengths')
```

```python
(jobs['skills_desc'].dropna().apply(len) < 1000).sum()
```

Explore Updated Experience Level

```python
exp_dist = jobs['formatted_experience_level'].value_counts()
exp_dist
```

```python
print("Percent of jobs with experience level specified:", exp_dist.sum() / len(jobs) * 100)
```

```python
exp_dist.plot(kind='bar')
plt.xlabel('Experience Level')
plt.ylabel('Number of Jobs')
plt.title('Distribution of Jobs by Experience Level')
```

```python
exp_dist.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Distribution of Jobs by Experience Level')
```

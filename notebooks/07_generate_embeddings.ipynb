```python
%pip install sentence-transformers
%pip install torch
%pip install numpy
from sentence_transformers import SentenceTransformer
import pandas as pd
import torch
import numpy as np
```

```python
print(torch.cuda.is_available())
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2',device='cpu')
jobs = pd.read_csv('data/processed/jobs_preprocessed.csv')
```

```python
jobs['embedding_text'] = jobs['title_clean'].fillna('') + jobs['description_clean'].fillna('') + jobs['skills_clean'].fillna('')
```

```python
embeddings = model.encode(
    jobs['embedding_text'].tolist(),
    batch_size=64,
    show_progress_bar=True
)
```

```python
np.save("data/processed/job_embeddings.npy", embeddings)
```
